```python
import matplotlib.pyplot as plt
%matplotlib inline

import os
import pandas as pd
from glob import glob
```

```python
DATA_DIR = 'data'
mask_files = glob(os.path.join(DATA_DIR, 'masks', '*.png'))
train_img_files = [os.path.join(DATA_DIR, 'images', os.path.basename(f))
                   for f in mask_files]
```

```python
fig, axes = plt.subplots(2, 5, figsize=(25, 10))
for i in range(5):
    img = plt.imread(train_img_files[i])
    print(img.shape)
    mask = plt.imread(mask_files[i])
    axes[0][i].imshow(img)
    axes[1][i].imshow(mask, cmap='gray')
```

```python
depth_file = os.path.join(DATA_DIR, 'depths.csv')
data = pd.read_csv(depth_file, index_col=0)
trains_ids = [os.path.basename(f).split(".png")[0] for f in mask_files]
z_values = data.loc[trains_ids]['z']
plt.hist(z_values)
plt.show()
```
